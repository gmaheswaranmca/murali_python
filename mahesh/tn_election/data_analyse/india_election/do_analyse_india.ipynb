### Imports and Config
* import the module csv
* define "csv file" as const "SOURCE_FILE"

```python
import csv

SOURCE_FILE : str = './parliament_edited.csv'
```

### Read Election Data from csv file - define function
* Test read data
  * Number of rows
  * display first two rows

```python
def read_election_data(source):
    with open(source, "r", encoding="utf-8") as file:
        reader = csv.DictReader(file)
        election_result = list(reader)
        return election_result
    
election_result = read_election_data(SOURCE_FILE)

print('Number of rows:', len(election_result))
print(election_result[0])
print(election_result[1])
```

# Year and its number of contituencies election conducted
* Define a function "count_consituencies_group_by_year:
  * takes rows as input
  * returns dict {"year" : "number of constituencies"}

```python
def count_consituencies_group_by_year(rows):
    constituency_count = {}
    for row in rows:
        row_rank = int(row['rank'])
        if row_rank != 1:
            continue 
        #
        row_year = int(row['year'])
        constituency_count[row_year] = constituency_count.get(row_year, 0) + 1
    #
    return constituency_count
#

def display_party_wise_winners_count(election_year_cosntituencies):
    print('\nElection Year wise constituencies count:')
    for year in election_year_cosntituencies:
        count = election_year_cosntituencies[year]
        print(f'{year} : {count}')
    #
#
election_year_cosntituencies = count_consituencies_group_by_year(election_result)
display_party_wise_winners_count(election_year_cosntituencies)
```

### Display head 5 rows
* Define a function "display_head"
  * takes rows as input
  * displays head 5 rows

```python
def display_head(rows):
    print('\nHead Rows:')
    for I in range(0,5):
        print(rows[I])
    #
#


display_head(election_result)
```

### Display tail 5 rows
* Define a function "display_tail"
  * takes rows as input
  * displays tail 5 rows

```python
def display_tail(rows):
    print('\nTail Rows:')
    for I in range(-1,-6,-1):
        print(rows[I])
        #
    #
#


display_tail(election_result)
```

### Read for election result 
* for year 2026 and 
* display head rows
* display tail rows


```python
def read_for_year(rows, year):
    year_rows = []
    for row in rows:
        row_year = int(row['year'])
        if row_year != year:
            continue
        #
        year_rows.append(row)
    #
    return year_rows
#

def sort_dict_by_val(rows):
    sorted_rows = dict(sorted(rows.items(), key=lambda pair:pair[1], reverse=True))
    return sorted_rows
#

election_result_2024 = read_for_year(election_result, 2024)
display_head(election_result_2024)
display_tail(election_result_2024)
```

### Group by party 
* Count the winners for each party
* Define the function "count_winner_group_by_party"
  * takes rows of election_results of the specific year
  * returns dict {'party' : 'winner count'}
* Display the party wise winners count
  * Define the function "display_party_wise_winners_count"
    * display as below the format 
      * party : count 
        * for each party
      * Example:
        * TVK : 5
        * DMK : 4
  * Display the party wise winners count for year 2026

```python
def count_winner_group_by_party(rows):
    party_winners_count = {}
    for row in rows:
        row_rank = int(row['rank'])
        if row_rank != 1:
            continue
        #
        row_party = row['party']
        party_winners_count[row_party] = party_winners_count.get(row_party, 0) + 1
    return party_winners_count

def display_party_wise_winners_count(party_winners_count):
    print('\nParty wise winners:')
    for party in party_winners_count:
        count = party_winners_count[party]
        print(f'{party} : {count}')

party_winners_count_2024 = count_winner_group_by_party(election_result_2024)
display_party_wise_winners_count(sort_dict_by_val(party_winners_count_2024))
```
