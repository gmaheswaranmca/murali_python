```python
import csv

SOURCE_FILE = './../data/election/tamil_nadu_edited.csv'
```

```python
def get_tn_election_data(source):
    with open(source, "r", encoding="utf-8") as file:
        reader = csv.DictReader(file)
        election_result = list(reader)
        return election_result

election_result = get_tn_election_data(SOURCE_FILE)

print('Results:')
election_summary = dict()
for row in election_result:
    # print(row)
    row_year = int(row["year"])
    row_rank = int(row["rank"])
    if row_rank == 1:
        election_summary[row_year] = election_summary.get(row_year, 0) + 1

print('Number of rows:', len(election_result))
for year in election_summary:
    print(f'\t{year} constituencies count:{election_summary[year]}')
```

```python
def get_tn_election_data_for_year(year):
    # Filter rows where YEAR == year
    filtered_data = []
    party_holdings = {}
    for row in election_result:
        row_year = int(row["year"])      
        
        if row_year == int(year):
            row_rank = int(row['rank'])
            row_party = row['party']
            row_rank_party = (row_party, row_rank)
            row_holdings = party_holdings.get(row_rank_party, 0) + 1
            party_holdings[row_rank_party] = row_holdings
            filtered_data.append(row)

    return filtered_data, party_holdings

year_2021_result, party_holdings = get_tn_election_data_for_year(2026)
print('Results:')
#for row in year_2021_result:
#    print(row)
print('Number of rows:', len(year_2021_result))

holdings_group_by_party_rank = {}
for key in party_holdings:
    rank = key[1]
    party = key[0]  
    holdings = party_holdings[key]

    if party in ('ADMK', 'AMMK', 'PMK', 'BJP'):
        pass
    elif party in ('DMK', 'INC', 'CPM', 'CPI', 'VCK', 'IUML', 'DMDK'):
        pass
    elif party in ('TVK'):
        pass
    else: 
        continue 

    if rank > 4:
        continue

    holdings_dict = holdings_group_by_party_rank.get(party, dict())
    holdings_dict[rank] = holdings_dict.get(rank, holdings)

    holdings_group_by_party_rank[party] = dict(sorted(holdings_dict.items()))

holdings_group_by_party_rank = dict(sorted(holdings_group_by_party_rank.items(), key=lambda row:row[1][1], reverse=True))
# sorted_holdings = sorted(sorted_holdings, key=lambda row:row[1], reverse=True)

for party in holdings_group_by_party_rank:
    holdings_dict = holdings_group_by_party_rank[party]    
    print(f'\t{party} : {holdings_dict}')
```

```python
def get_tn_election_data_for_party(party):
    filtered_data = [row for row in year_2021_result if row["party"] == party]
    return filtered_data

year_2021_result_tvk = get_tn_election_data_for_party('TVK')
print('Results:')
for row in year_2021_result_tvk:
    print(row)
print('Number of rows:', len(year_2021_result_tvk))
```
