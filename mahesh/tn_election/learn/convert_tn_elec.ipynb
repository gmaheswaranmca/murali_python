```python
import csv

SOURCE_FILE = './../data/election/tamil_nadu.csv'
DEST_FILE = './../data/election/tamil_nadu_edited.csv'

HEADER = ['constituency', 'party', 'year', 
    'state', 'constituency_number', 'rank', 'constituency_type', 
    'candidate', 'sex', 'age', 'category', 
    'votes_received', 'votes_polled', 'color', 'district']
def read_tn_election_data(source):
    with open(file = source, mode = "r", encoding="utf-8") as file:
        reader = csv.reader(file)
        next(reader) # to skip the header row, otherwise header will also include 
        election_result = list(reader) # 'list of rows', where each row is also of the type 'list'
        return election_result # 'rows without header'    

def write_tn_election_data(election_result, header, dest):
    with open(file = dest, mode = "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

        writer.writerow(header)
        writer.writerows(election_result)

election_result = read_tn_election_data(SOURCE_FILE)
print(type(election_result[0]))
print(election_result[0])
write_tn_election_data(election_result, HEADER, DEST_FILE)
```
