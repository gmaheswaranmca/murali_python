```python
import csv

SOURCE_FILE = './../data/election/parliament.csv'
DEST_FILE = './../data/election/parliament_edited.csv'

HEADER = ['state_code', 'constituency_number', 'original_serial_number', 
    'candidate_serial_number', 'candidate', 'political_alliance', 
    'party', 'color', 'evm_votes', 
    'postal_votes', 'migrant_votes', 'votes_received', 
    'vote_percentage', 'votes_polled', 'deposit_gained', 
    'party_full_name', 'state', 'constituency', 
    'reservation_type', 'election_phase', 'polling_date', 
    'nomination_filing_date', 'vote_percentage_precise', 'candidate_photo_url', 
    'year', 'rank']
def read_tn_election_data(source):
    with open(file = source, mode = "r", encoding="utf-8") as file:
        reader = csv.reader(file)
        next(reader) # to skip the header row, otherwise header will also include 
        election_result = list(reader) # 'list of rows', where each row is also of the type 'list'
        return election_result # 'rows without header'    

def create_index(rows, index_keys=tuple()):
    index_table = {}
    for index, row in enumerate(rows):
        '''if index == 0:
            continue
        #'''
        index_values = tuple([row[key] for key in index_keys])
        #
        index_rows = index_table.get(index_values, list())
        index_rows.append((index, row[11]))
        index_table[index_values] = index_rows
    #
    return index_table
#

def update_rank(rows, index_table):
    for J, index_key in enumerate(index_table):
        index_val = index_table[index_key] 
        if J ==0:
            print('<',index_key,':',index_val, '>')
        index_val = sorted(index_val, key=lambda t:int(t[1]), reverse=True)
        if J ==0:
            print('after', '<',index_key,':',index_val, '>')
        for rank, indexTuple in enumerate(index_val):
            index, votesReceived = indexTuple 
            rows[index].append(rank+1) 
        #

def write_tn_election_data(election_result, header, dest):
    with open(file = dest, mode = "w", newline="", encoding="utf-8") as file:
        writer = csv.writer(file)

        writer.writerow(header)
        writer.writerows(election_result)
    #
#

election_result = read_tn_election_data(SOURCE_FILE)
index_table = create_index(election_result, (24,0,1))
update_rank(election_result, index_table)
print(type(election_result[0]))
print(election_result[0])
print(election_result[1])

print(len(election_result[0]))
print(len(election_result[1]))


write_tn_election_data(election_result, HEADER, DEST_FILE)
```
