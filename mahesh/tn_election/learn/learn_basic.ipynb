### Transform 
* list of candidates dict given
into 
* list of dict
  * {constituency, party, candidate, rank, votes_polled, votes_received, votes_per, is_mla, mla_str}
### Display transformed candidates as
| Constituency | Candiate | Polled Votes# | Received Votes# | Vote% | MLA Status |
* Display who has secured highest vote % received out of all
  * Note: Print vote % also
### Write four functions 
* 1 - to transform
* 2 - display table of candidates
* 3 - find largest vote% getter
* 4 - display largest vote% getter
* Main / Driver code to call above functions

```python
def transform_candiates(candidates):
    result_candidates = []
    for row in tn_election_result:
        constituency : str = row['constituency']
        party : str = row['party']
        candidate : str = row['candidate']
        rank : int = int(row['rank'])
        votes_polled : int =  int(row['votes_polled'])
        votes_received : int = int(row['votes_received'])

        # Calculate 'Vote Percentage'
        votes_per : float = (votes_received / votes_polled) * 100
        is_mla : bool = (rank == 1) 

        if is_mla:
            mla_str: str = f'MLA'
        else:
            mla_str: str = f'Not MLA'
        #
        row['votes_per'] = votes_per
        row['is_mla'] = is_mla
        row['mla_str'] = mla_str

        candidate_row = {
            'constituency' : constituency,
            'party' : party,
            'candidate' : candidate,
            'rank' : rank,
            'votes_polled' : votes_polled,
            'votes_received' : votes_received,
            'votes_per' : votes_per,
            'is_mla' : is_mla,
            'mla_str' : mla_str
        }

        result_candidates.append(candidate_row)
    #
    return result_candidates
#

def display_candidates(candiates):
    for row in candiates:
        print(f'{row["constituency"]:20} | {row["candidate"]:20} | {row["votes_polled"]:8} | {row["votes_received"]:8} | {row["votes_per"]:10.2f} | {row["mla_str"]:10}')
    #
#

def find_largest_vote_per_getter(candidates):
    max_vote_per_candidate = max(candidates, key=lambda row:row['votes_per'])
    return max_vote_per_candidate
#

def display_largest_vote_per_getter(candiate):
    print(f'Note:{candiate["candidate"]} is highest percentage {candiate["votes_per"]:.2f}% votes received candiate')
#

# Main Code
transformed_candiates = transform_candiates(tn_election_result)
display_candidates(transformed_candiates)
max_vote_per_candidate = find_largest_vote_per_getter(transformed_candiates)
display_largest_vote_per_getter(max_vote_per_candidate)
```
